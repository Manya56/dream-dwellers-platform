```python
import pandas as pd
df=pd.read_csv(r"C:\Users\akhya\Downloads\PJME_hourly.csv.zip")
df['Datetime']=pd.to_datetime(df['Datetime'])
df.set_index('Datetime',inplace=True)
print(df.head())
print(df.info())
```

```python
def create_features(df):
    df = df.copy()
    df['hour'] = df.index.hour
    # Use ['day'] because isocalendar() returns 'year', 'week', and 'day'
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    
    df['lag_24h'] = df['PJME_MW'].shift(24)
    df['lag_7d'] = df['PJME_MW'].shift(24 * 7)
    
    df.dropna(inplace=True)
    return df

# Run the function
df_featured = df_featured.sort_index()
print(df_featured.head())
```

```python
split_date = '2018-07-01'

# Changed df.featured to df_featured
train = df_featured.loc[df_featured.index < split_date].copy()
test = df_featured.loc[df_featured.index >= split_date].copy()

features = ['hour', 'dayofweek', 'quarter', 'month', 'year', 'dayofyear', 'lag_24h', 'lag_7d']
# Changed 'consumption' to the actual column name 'PJME_MW'
target = 'PJME_MW'

X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]
```

```python
import xgboost as xgb
model=xgb.XGBRegressor(
    n_estimators=1000,
    early_stopping_rounds=50,
    learning_rate=0.03,
    max_depth=6
)
model.fit(
    X_train,y_train,
    eval_set=[(X_train,y_train),(X_test,y_test)],
    verbose=100
    )
model.save_model(r"C:\Users\akhya\OneDrive\Documents\mlproj\model.json")
print("Model saved successfully!")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
test['prediction']=model.predict(X_test)
mae=mean_absolute_error(test['PJME_MW'],test['prediction'])
rmse=np.sqrt(mean_squared_error(test['PJME_MW'],test['prediction']))
print(f"Mean absolute error:{mae:.2f} kWh")
print(f"Root Mean Squared Error:{rmse:.2f} kWh")
```
